# Intro a Análisis Exploratorio de Datos con Pandas

```python
# Cargar en una variable la URL del archivo csv de Github
url = "https://raw.githubusercontent.com/pandas-dev/pandas/master/doc/data/titanic.csv"
```

```python
# Importar la librería Pandas
import pandas as pd
```

```python
# Usar el método read_csv de Pandas (pd) que recibe como argumento el archivo csv y retorna un dataframe
df = pd.read_csv(url)
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(6,4))
sns.countplot(x='Pclass', data=df, palette='Set1')
plt.title("Distribucion de la clase")
plt.xlabel("Clase")
plt.ylabel("Cantidad")
plt.show()
```
